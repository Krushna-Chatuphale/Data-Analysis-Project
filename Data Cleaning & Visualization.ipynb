```python
# Task 1: Load the Data

# Step 1: Import the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

print("First 5 rows:")
print(df.head())

print("Shape of dataset:", df.shape)

print("Column names:", df.columns)
```

```python
# Task 2: Check for Missing or Duplicate Data

print("Missing values in each column:")
print(df.isnull().sum())

df.ffill(inplace=True)

print("Number of duplicate rows:", df.duplicated().sum())

df.drop_duplicates(inplace=True)

print("New shape after removing duplicates:", df.shape)
```

```python
# Task 3: Summary and Basic Stats

# Summary statistics
print("Summary statistics:")
print(df.describe())

print("Data info:")
print(df.info())

print("Species count:")
print(df['species'].value_counts())

print("Average sepal length:", df['sepal_length'].mean())
print("Average petal width:", df['petal_width'].mean())

print("Rows where sepal_length > 6:")
print(df[df['sepal_length'] > 6])
```

```python
# Task 4: Bar Chart Visualization

# Bar chart for species count
sns.countplot(x='species', data=df)
plt.title("Species Count")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
# Task 5: Histogram and Line Plot

# Histogram of sepal_length
df['sepal_length'].plot(kind='hist', bins=15, color='skyblue')
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.show()

# Line plot of sepal_length and sepal_width
plt.plot(df['sepal_length'], label='Sepal Length', color='blue')
plt.plot(df['sepal_width'], label='Sepal Width', color='orange')
plt.title("Sepal Length and Width Line Plot")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Make sure df is loaded, for example:
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")

# Filter numeric columns
numeric_df = df.select_dtypes(include='number')

# Plot correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
